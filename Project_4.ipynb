##  Problem 1:
### There are various stocks for which we have collected a data set. Which all stocks are  apparently similar in performance?

```python
# Dropping the columns date and SP500
df = data.drop(['DATE','SP500'],axis=1)
df.head()
```

```python
# To know which all stocks are similar in performance we find the correlation among them 
corr = df.corr()
```

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
# Finding the top 20 highly correlated columns and printing them
def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations\n","-"*50)
print(get_top_abs_correlations(df, 20))
```

## Problem 2:
### How many Unique patterns that exist in the historical stock data set, based on fluctuations in price ?

```python
# To find the unique patterns based on the fluctuations in price, we need to use K-Means Clustering algorithm
# Using the Elbow Curve plot we can determine the number of unique patterns existing in the given data.
```

```python
from math import sqrt
from sklearn.cluster import KMeans
#Calculate average annual percentage return and volatilities over a theoretical one year period
returns = df.pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = df.pct_change().std() * sqrt(252)

#format the data as a numpy array to feed into the K-Means algorithm
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T

X = data
distorsions = []
for k in range(2, 20):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distorsions.append(k_means.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

```python
# From the graph, we can infer that the ideal number of clusters should be around 3-5
```

### Therefore, in the given historical stock data around 3-5 unique patterns exist based on fluctuations in price

## Problem 3:
### Identify which all stocks are moving together and which all stocks are different from each other.

```python
# K-Means Clustering is performed to identify which all stocks are moving together and which all are different from each other

from scipy.cluster.vq import kmeans,vq
# computing K-Means with K = 5 (5 clusters)
centroids,_ = kmeans(data,5)
# assign each sample to a cluster
idx,_ = vq(data,centroids)

pylab.rcParams['figure.figsize'] = (8,6)
# some plotting using numpy's logical indexing
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()
```
